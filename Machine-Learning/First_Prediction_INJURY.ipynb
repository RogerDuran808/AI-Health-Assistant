```python
# Importar les llibreries necessàries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

```

```python
#Carreguem el dataset
df = pd.read_csv('injury_data.csv')
```

```python
df.head()
```

```python

# La columna 'injury' és la variable objectiu (target).
X = df.drop('Likelihood_of_Injury', axis=1)
y = df['Likelihood_of_Injury']

print(X.head())
print("\n", y.head())
```

```python

# Dividir les dades en entrenament (80%) i test (20%) amb estratificació per mantenir la distribució de classes
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Estandaritzem les dades. CREC QUE AQUEST PART ESTA MALAMENTE JA QUE NO FA ENTRE -1 I 1
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print(X_train)

Xtraindata = pd.DataFrame(X_train, columns=X.columns)
print("\n")
print(Xtraindata.head())
```

```python
# Entrenar un model Random Forest
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Realitzar prediccions sobre el conjunt de test i avaluar el model
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred), "\n")
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
